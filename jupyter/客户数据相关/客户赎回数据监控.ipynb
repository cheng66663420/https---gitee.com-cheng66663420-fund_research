```python
import pandas as pd
file_path = 'D:/BaiduNetdiskWorkspace/1-基金投研/5-data/客户交易数据/'

port_stats = pd.read_excel(
    file_path + '基金投顾业务收益率.xlsx',
    sheet_name='报表1_组合销售情况'
)

port_stats['基金投顾产品类型'] = port_stats['基金投顾产品类型'].fillna(method='ffill')
port_stats['所属一级分公司'] = port_stats['所属一级分公司'].fillna(method='ffill')
port_stats['所属营业部'] = port_stats['所属营业部'].fillna(method='ffill')
port_stats['所属员工'] = port_stats['所属员工'].fillna(method='ffill')
port_stats['客户编号'] = port_stats['客户编号'].apply(lambda s: format(s, '.0f'))

# 每个组合本周赎回的前5名分公司
redeem_total = (
    port_stats.groupby(by=['基金投顾产品类型', '所属一级分公司'])[
        ['本周基金投顾赎回金额(万)', '本月基金投顾赎回金额(万)', '本年基金投顾赎回金额(万)', '累计基金投顾赎回金额(万)']
    ]
        .sum()
        .reset_index()
        .sort_values(['本周基金投顾赎回金额(万)','基金投顾产品类型', '所属一级分公司',], ascending=False)
)
```

本周个策略赎回前5的分公司

```python
redeem_total.groupby(by=['基金投顾产品类型']).head(5).sort_values(by=['基金投顾产品类型', '本周基金投顾赎回金额(万)'], ascending=False)
```

本周个策略赎回前5的客户

```python
# 每个组合当周赎回最多的5个客户
port_stats.groupby(by=['基金投顾产品类型']).apply(
    lambda s: s.sort_values(['本周基金投顾赎回金额(万)'], ascending=False).head(5)[
        [
            '所属一级分公司','所属营业部','所属员工', '客户编号', '本周基金投顾赎回金额(万)',
            '本月基金投顾赎回金额(万)', '本年基金投顾赎回金额(万)', '累计基金投顾赎回金额(万)'
        ]
    ]
)
```

```python

company_stats = pd.read_excel(
    file_path + '基金投顾业务收益率.xlsx', 
    sheet_name='Tab组件-分公司客户购买情况'
)

company_stats['所属一级分公司'] = company_stats['所属一级分公司'].fillna(method='ffill')
company_stats['所属营业部'] = company_stats['所属营业部'].fillna(method='ffill')
company_stats['所属员工'] = company_stats['所属员工'].fillna(method='ffill')
company_stats['客户编号'] = company_stats['客户编号'].apply(lambda s: format(s, '.0f'))
```
