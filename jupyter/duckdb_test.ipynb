```python
def get_data(date:str) ->pl.DataFrame:
    query = f"""
        SELECT 
        * 
        FROM 
        fund_derivatives_fund_alpha_performance 
        WHERE END_DATE = '{date}'
    """
    return pl.read_database_uri(query, JJTG_URI).select(pl.all().exclude("ID"))

```
