```python
import plotly.graph_objects as go

# 数据
start_date = "2023-03-17"
end_date = "2024-07-12"
cumulative_rate_of_return = 2.92
annualized_rate_of_return = 2.21
annualized_volatility = 0.11

indicators = go.Figure()
indicators.add_trace(go.Indicator(
    mode="number+delta",
    value=cumulative_rate_of_return,
    title={'text':"累计收益率", 'font': {'size': 24, 'family': 'Arial Black', 'color': "black"}},
    number = {'suffix': "%", 'valueformat': '.2f'},
    domain={'row': 0, 'column': 0},

))
indicators.add_trace(go.Indicator(
    mode="number+delta",
    value=annualized_rate_of_return,
    title=dict(text="年化收益率"),
    number = {'suffix': "%"},
    domain={'row': 1, 'column': 0},
   
))
indicators.add_trace(go.Indicator(
    mode="number+delta",
    value=annualized_volatility,
    title=dict(text="年化波动率"),
    number = {'suffix': "%"},
    domain={'row': 2, 'column': 0},

))
# 创建仪表盘对象

indicators.update_layout(
    grid = {'rows': 3, 'columns': 1, 'pattern': "independent"},
    margin=dict(l=20, r=20, t=50, b=50),
    paper_bgcolor="#999999"
)
# 显示图形
indicators.show()
```

```python
from sqlalchemy import create_engine
import pandas as pd
from quant_utils.constant import DB_CONFIG
DATABASE_URI = f'mysql+pymysql://{DB_CONFIG["jjtg"]["user"]}:{DB_CONFIG["jjtg"]["pwd"]}@{DB_CONFIG["jjtg"]["host"]}/{DB_CONFIG["jjtg"]["database"]}'

JJTG_ENGINE = create_engine(
    DATABASE_URI,
    max_overflow=10,
    pool_recycle=3600,
    pool_size=10,
)
query_sql = """
SELECT
  a.TICKER_SYMBOL,
  b.LEVEL_1,
  b.LEVEL_2,
  b.LEVEL_3,
  a.RETURN_RATE_YTD as YTD
FROM
  fund_nav_gr a
  JOIN fund_type_own b ON a.TICKER_SYMBOL = b.TICKER_SYMBOL 
WHERE
  1 = 1 
  AND b.REPORT_DATE = '20240331' 
  AND b.LEVEL_1 = '固收' 
  AND a.END_DATE = '20240712'
"""
df = pd.read_sql_query(query_sql, JJTG_ENGINE)
```

```python
df["基金类别"] = df["LEVEL_1"] + "-" + df["LEVEL_2"] + "-" + df["LEVEL_3"]
```

```python
df.sort_values(by="基金类别", ascending=False, inplace=True)
df = df[df["YTD"]<= 50]
```

```python
df.groupby(by="基金类别").describe()
```

```python
import plotly.express as px  # 1、px实现
fig = px.box(
    df,
    x='基金类别',  # 星期
    y='YTD',
    color= '基金类别',
    points=False
)

fig.show()
```
